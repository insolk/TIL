```python
%matplotlib inline
import pandas as pd
import matplotlib as plt
```

```python
##1. csv 파일 불러오기 (convenient_store.csv)
con = pd.read_csv('convenient_store.csv')
```

```python
##2. 전체 컬럼 정보, null 값 유무 확인
con.info()
```

```python
##3. 개수, 평균, 편차, 최소, 최대값 확인
# print(con['hourly_wage'].count())
# print(con['hourly_wage'].mean())
# print(con['hourly_wage'].std())
# print(con['hourly_wage'].min())
# print(con['hourly_wage'].max())
con.describe()
```

```python
##4. 지역에 대한 통계, 개수, 유니크한 정보, 제일 빈도가 높은 지역
# con['area'].unique()
# con['area1'].value_counts()
con.area.describe()
```

```python
##5. 시간 당 급여가 6500원 이상인 지역의 편의점 정보 출력 (상위 10개만)
# con[con['hourly_wage']>=6500].head(10)
high_wage = con[con.hourly_wage >=6500]
high_wage
```

```python
##6. 시간 당 급여가 높은 순서로 정렬 (sort_value() 함수 사용, 상위 10개만 출력)
# con.sort_values(by='hourly_wage', ascending=False).head(10)
high_wage.sort_values(by='hourly_wage', ascending=0).head(10)
```

```python
##7. 영등포구에서 시간 당 급여가 6000원 이상인 편의점 검색
# con[(con['area1']=='영등포구')&(con['hourly_wage']>=6000)].head(10)
y_high_wage = con[(con.hourly_wage>=6000) & (con.area1 == '영등포구')]
y_high_wage
```

```python
##8. CU 편이점만 출력 (상위 10개만)
#con[con['name']=='CU'].head(10)
cu = con[con.company.str.contains('CU')]
cu.head(10)
```

```python
##9. 지역 컬럼(location)을 추가한 다음, in Seoul 이라는 값 저장, 상위 5개 출력
con['location'] = "in Seoul"
con.head(5)
```

```python
##10. 6000원 이상 컬럼 추가(more_than_6000) -> True, False 값 저장 (상위 20개 출력)
# con['more_than_6000'] = con['hourly_wage']>=6000
con['more_than_6000'] = con['hourly_wage']>6000
con.head(20)
```

```python
##11. more_than_6000 컬럼에서 True인 데이터들의 평균, 개수, 편차 등의 정보 출력
# print(con[con['more_than_6000']==True]['hourly_wage'].mean())
# print(con[con['more_than_6000']==True]['hourly_wage'].count())
# print(con[con['more_than_6000']==True]['hourly_wage'].std())
con[con['more_than_6000']==True].describe()
```

```python
##12. more_than_6000 이름의 함수를 생성하고, 6000원이상인 경우 A group, 아니면 B group을 반환하는 함수 생성
# def more_than_6000(x: bool) -> str:
#     if x == True:
#         return 'A group'
#     else:
#         return 'B group'
    
def more_than_6000(x: int) -> int:
    if x >= 6000:
        return 'A group'
    else:
        return 'B group'
```

```python
##13. more_than_6000_f 컬럼 생성하고 more_than_6000 함수의 결과를 저장
# con['more_than_6000_f']=con['more_than_6000'].apply(lambda x : more_than_6000(x))
con['more_than_6000_f']=con.hourly_wage.map(lambda x: more_than_6000(x))
```

```python
##14. 지금까지의 결과 상위 10개를 출력
con.head(10)
```

```python
##15-1. more_than_6000가 True인 데이터의 지역과 시간당 급여를 가진 새로운 데이터프레임 생성(data2)
# data2 = con[con['more_than_6000'] == True]
data2 = con[con['more_than_6000'] == True][['area1', 'hourly_wage']]
##15-2. data2 데이터를 시간당 급여 순으로 정렬 (높은순)
data2.sort_values(by='hourly_wage', ascending=False)
```

```python
##16. data2를 darta2.csv 파일로 저장
# data2.to_csv("data2.csv", mode='w')
data2.to_csv("data2.csv", mode='w', index =False)
```

```python
##17. 시간당 급여를 histogram 으로 표시 (matplotlib hist() 사용)
# import matplotlib.pyplot as plt
# import seaborn as sns
# plt.hist(con['hourly_wage'])
con.hourly_wage.hist(bins=10)
```

```python
##18. 시간당 급여를 box 차트로 표시 
# figure,axes1= plt.subplots(nrows=1, ncols =1)
# figure.set_size_inches(18, 12)
# sns.barplot(x='name', y='hourly_wage', data=con, ax =axes1)
con.boxplot(column = 'hourly_wage', return_type='dict')
```

```python
##19. 시간당 급여를 box 차트로 표시(이름순으로)
# figure,axes1= plt.subplots()
# figure.set_size_inches(18, 12)
# con1 = con.sort_values(by='name')
# sns.barplot(x='name', y='hourly_wage', data=con1, ax =axes1)
con.boxplot(column = 'hourly_wage',by='name')
```

```python
##19. 시간당 급여를 box 차트로 표시(지역순으로)
# figure,axes1= plt.subplots()
# figure.set_size_inches(100, 12)
# con2 = con.sort_values(by='area')
# sns.barplot(x='area', y='hourly_wage', data=con2, ax =axes1)
con.boxplot(column = 'hourly_wage',by='area1')
```
